```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np

ROOT_DIR = os.path.abspath("")
```

# Histogram Equalization

```python
IMAGE_DIR = os.path.join(ROOT_DIR, "")

gray_image = cv2.imread("bina.png")
gray_image = cv2.cvtColor(gray_image, cv2.COLOR_BGR2GRAY)

equ = cv2.equalizeHist(gray_image)
plt.imshow(equ)
```

# GrayScale Histogram Equalization

```python
gray_img_eqhist=cv2.equalizeHist(gray_image)
gray_img1_eqhist=cv2.equalizeHist(gray_image)
hist=cv2.calcHist(gray_img_eqhist,[0],None,[256],[0,256])
hist1=cv2.calcHist(gray_img1_eqhist,[0],None,[256],[0,256])
plt.subplot(121)
plt.plot(hist)
plt.subplot(122)
plt.plot(hist1)
plt.show()
```

```python
eqhist_images=np.concatenate((gray_img_eqhist,gray_img1_eqhist),axis=1)
cv2.imshow("Images",eqhist_images) #siyah beyaz hali
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Erode Function

```python
kernel = np.ones((3,3),np.uint8)
output2 = cv2.erode(gray_image,kernel,iterations = 3)
plt.imshow(output2)
plt.show()
```

# Canny Funciton

```python
img = cv2.imread(r"C:\Users\elif.ayanoglu\Downloads\bina.png")
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
(thresh, output2) = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)
output2 = cv2.Canny(output2, 180, 255,apertureSize=7,L2gradient = True)
plt.imshow(output2)
plt.show()
```

```python
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 800, 255, cv2.THRESH_BINARY)
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 160, 255, cv2.THRESH_BINARY)
(thresh, blackAndWhiteImage) = cv2.threshold(gray_image, 200, 255, cv2.THRESH_BINARY)
plt.imshow(blackAndWhiteImage)
plt.show()

cv2.imshow("Images",blackAndWhiteImage)#siyah beyaz versiyonu iin
cv2.waitKey(0)
cv2.destroyAllWindows()
```

# Clahe

```python
clahe=cv2.createCLAHE(clipLimit=40)
gray_img_clahe=clahe.apply(gray_img_eqhist)
gray_img1_clahe=clahe.apply(gray_img1_eqhist)
images=np.concatenate((gray_img_clahe,gray_img1_clahe),axis=1)

cv2.imwrite("Image1.jpg",images)
plt.imshow(images)

cv2.imshow("Images",images)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
th=80
max_val=255
ret, o1 = cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_BINARY)
cv2.putText(o1,"Thresh_Binary",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret, o2 = cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_BINARY_INV)
cv2.putText(o2,"Thresh_Binary_inv",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)
ret, o3 = cv2.threshold(gray_img_clahe, th, max_val, cv2.THRESH_TOZERO)
cv2.putText(o3,"Thresh_Tozero",(40,100),cv2.FONT_HERSHEY_SIMPLEX,2,(255,255,255),3,cv2.LINE_AA)


final=np.concatenate((o1,o2,o3),axis=1)


cv2.imwrite("Image1.jpg",final)
plt.imshow(final)

cv2.imshow("Images",images)#siyah beyaz grnts iin
cv2.waitKey(0)
cv2.destroyAllWindows()
```
